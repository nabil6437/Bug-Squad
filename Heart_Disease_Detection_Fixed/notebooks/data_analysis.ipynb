# Heart Disease — Data Analysis & Visualization

This notebook covers:
1. Loading and preprocessing the raw dataset
2. Statistical summary
3. Correlation heatmap
4. Feature distributions (histograms & boxplots)
5. Feature importance (correlation with target)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
matplotlib .use ('Agg')
import matplotlib .pyplot as plt
import sys ,os

sys .path .insert (0 ,os .path .join (os .getcwd (),'..'))
from utils .data_processing import run_pipeline ,feature_selection

RAW =os .path .join ('..','data','raw_data.csv')
CLEANED =os .path .join ('..','data','cleaned_data.csv')
REPORT_DIR =os .path .join ('..','reports')
os .makedirs (REPORT_DIR ,exist_ok =True )
```

## 1. Load & Preprocess Data

```python

df_clean ,df_scaled ,scaler ,encoder =run_pipeline (RAW ,CLEANED ,scale =True )

print (f"Clean shape  : {df_clean .shape }")
print (f"Scaled shape : {df_scaled .shape }")
print (f"\nClass distribution:\n{df_clean ['target'].value_counts ()}")
```

## 2. Statistical Summary

```python
print ('=== Statistical Summary (Clean Data) ===')
df_clean .describe ()
```

## 3. Correlation Heatmap

```python
fig ,ax =plt .subplots (figsize =(12 ,10 ))
sns .heatmap (df_clean .corr (),annot =True ,fmt ='.2f',cmap ='coolwarm',ax =ax )
ax .set_title ('Feature Correlation Heatmap (Raw Features)',fontsize =14 ,pad =12 )
plt .tight_layout ()
fig .savefig (os .path .join (REPORT_DIR ,'correlation_heatmap.png'),dpi =150 )
plt .show ()
print ('Saved → reports/correlation_heatmap.png')
```

## 4. Histograms & Boxplots

```python
numeric_cols =['age','trestbps','chol','thalach','oldpeak']

fig ,axes =plt .subplots (2 ,3 ,figsize =(15 ,10 ))
axes =axes .flatten ()
for i ,col in enumerate (numeric_cols ):
    df_clean [col ].hist (ax =axes [i ],bins =30 ,edgecolor ='black',color ='steelblue')
    axes [i ].set_title (f'Distribution of {col }')
for j in range (len (numeric_cols ),len (axes )):
    fig .delaxes (axes [j ])
plt .tight_layout ()
fig .savefig (os .path .join (REPORT_DIR ,'histograms.png'),dpi =150 )
plt .show ()
print ('Saved → reports/histograms.png')
```

```python

fig ,axes =plt .subplots (2 ,3 ,figsize =(15 ,10 ))
axes =axes .flatten ()
for i ,col in enumerate (numeric_cols ):
    df_clean .boxplot (column =col ,ax =axes [i ])
    axes [i ].set_title (f'Boxplot of {col }')
for j in range (len (numeric_cols ),len (axes )):
    fig .delaxes (axes [j ])
plt .tight_layout ()
fig .savefig (os .path .join (REPORT_DIR ,'boxplots.png'),dpi =150 )
plt .show ()
print ('Saved → reports/boxplots.png')
```

## 5. Feature Importance (Correlation with Target)

```python
selected_features ,corr_values =feature_selection (df_clean ,'target',threshold =0.05 )

fig ,ax =plt .subplots (figsize =(10 ,6 ))
corr_values .sort_values ().plot (kind ='barh',ax =ax ,color ='teal')
ax .set_title ('Feature Importance (Absolute Correlation with Target)')
ax .set_xlabel ('Absolute Correlation')
plt .tight_layout ()
fig .savefig (os .path .join (REPORT_DIR ,'feature_importance.png'),dpi =150 )
plt .show ()

print (f"\nSelected features (corr >= 0.05): {selected_features }")
print ('Saved → reports/feature_importance.png')
```
